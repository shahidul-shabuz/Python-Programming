In Python, inheritance is a fundamental concept of object-oriented programming (OOP) that allows you to create new classes (derived classes) based on existing classes (base classes). The derived class inherits attributes and behaviors from the base class, enabling code reuse and promoting a hierarchical structure.

Inheritance is a powerful feature of object-oriented programming that allows us to reuse code and create more complex classes. In Python, inheritance is implemented using the inherits keyword.

To define a derived class that inherits from a base class, you can use the following syntax:

```python
class BaseClass:
    # Base class attributes and methods

class DerivedClass(BaseClass):
    # Derived class attributes and methods
```

In this example, DerivedClass is the derived class, and BaseClass is the base class. The derived class inherits all the attributes and methods from the base class.

When an instance of the derived class is created, it can access both the attributes and methods defined in the derived class itself and those inherited from the base class. If a method is defined in both the derived and base class with the same name, the method in the derived class will override the one in the base class.

Here's a simple example to illustrate inheritance in Python:

```python
class Vehicle:
    def __init__(self, brand):
        self.brand = brand

    def drive(self):
        print(f"The {self.brand} is being driven.")

class Car(Vehicle):
    def honk(self):
        print("Beep beep!")

class Motorcycle(Vehicle):
    def wheelie(self):
        print("Performing a wheelie!")

car = Car("Toyota")
car.drive()  # Inherited from Vehicle class
car.honk()   # Defined in Car class

motorcycle = Motorcycle("Honda")
motorcycle.drive()   # Inherited from Vehicle class
motorcycle.wheelie() # Defined in Motorcycle class
```

In this example, Vehicle is the base class, and Car and Motorcycle are derived classes. Both derived classes inherit the drive() method from the Vehicle class. The Car class adds a new method honk(), while the Motorcycle class adds a new method wheelie(). When creating instances of these classes, you can see how the inherited methods can be called, as well as the specific methods defined in each derived class.

Inheritance allows you to create specialized classes that extend the functionality of the base class while maintaining code reuse and modularity. It is an essential concept in object-oriented programming.

```python
class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def drive(self):
        print("The car is driving.")

class SportsCar(Car):
    def __init__(self, make, model, year, top_speed):
        super().__init__(make, model, year)
        self.top_speed = top_speed

    def drive(self):
        print("The sports car is driving.")
        print("The top speed of the sports car is", self.top_speed)

car = Car("Honda", "Accord", 2023)
car.drive()

sports_car = SportsCar("Ferrari", "488 GTB", 2022, 211 mph)
sports_car.drive()
```

The Dog and Cat classes inherit all of the methods and properties of the Animal class. They also define their own methods and properties that are specific to dogs and cats.

## Single Inheritance

Single inheritance is a feature of object-oriented programming languages where a class can inherit properties and methods from a single parent class

single inheritance.JPG

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def eat(self):
        print(f"{self.name} is eating.")


class Dog(Animal):
    def bark(self):
        print(f"{self.name} is barking.")


# Creating an instance of the Dog class
my_dog = Dog("Buddy")

# Accessing the properties and methods of the Animal class inherited by Dog
print(my_dog.name)  # Output: Buddy
my_dog.eat()  # Output: Buddy is eating.

# Accessing the method specific to the Dog class
my_dog.bark()  # Output: Buddy is barking.
```

In the example above, we have two classes: Animal and Dog. The Dog class inherits from the Animal class using single inheritance. The Animal class has an __init__ method and an eat method. The Dog class adds a bark method.

When we create an instance of the Dog class (my_dog = Dog("Buddy")), we can access the properties and methods of the Animal class using the my_dog object. The Dog class inherits the name property from the Animal class and can call the eat method.

Additionally, the Dog class has its own method called bark, which is specific to the Dog class. We can call this method using the my_dog object.

# Hierarchical Inheritance

Hierarchical inheritance in Python refers to a scenario where a single parent class serves as the base for multiple child classes. Each child class inherits the attributes and methods of the parent class, while also having its own unique attributes and methods.

hierchical inheritance.JPG

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def eat(self):
        print(f"{self.name} is eating.")

    def sleep(self):
        print(f"{self.name} is sleeping.")


class Dog(Animal):
    def bark(self):
        print(f"{self.name} is barking.")


class Cat(Animal):
    def meow(self):
        print(f"{self.name} is meowing.")


# Create instances of the child classes
dog = Dog("Buddy")
cat = Cat("Whiskers")

# Accessing methods from the parent class
dog.eat()  # Output: Buddy is eating.
cat.sleep()  # Output: Whiskers is sleeping.

# Accessing methods unique to each child class
dog.bark()  # Output: Buddy is barking.
cat.meow()  # Output: Whiskers is meowing.
```

In this example, we have a parent class called Animal, which has common attributes and methods shared by all animals. The Dog class and Cat class inherit from the Animal class, thereby inheriting its attributes and methods.

The Dog class has an additional method bark, which is unique to dogs. The Cat class has an additional method meow, which is unique to cats.

We create instances of the child classes (dog and cat) and demonstrate how they can access the methods from the parent class (eat and sleep), as well as their own unique methods (bark and meow).

This is an example of hierarchical inheritance, where the Animal class acts as the base for the Dog and Cat classes.

```python
class Shape:
    def __init__(self, color):
        self.color = color

    def draw(self):
        print(f"Drawing a {self.color} shape.")


class Circle(Shape):
    def __init__(self, color, radius):
        super().__init__(color)
        self.radius = radius

    def calculate_area(self):
        area = 3.14 * self.radius**2
        print(f"The area of the {self.color} circle is {area}.")


class Square(Shape):
    def __init__(self, color, side):
        super().__init__(color)
        self.side = side

    def calculate_area(self):
        area = self.side**2
        print(f"The area of the {self.color} square is {area}.")


# Create instances of the child classes
circle = Circle("Red", 5)
square = Square("Blue", 7)

# Accessing methods from the parent class
circle.draw()  # Output: Drawing a Red shape.
square.draw()  # Output: Drawing a Blue shape.

# Accessing methods unique to each child class
circle.calculate_area()  # Output: The area of the Red circle is 78.5.
square.calculate_area()  # Output: The area of the Blue square is 49.
```

In this example, we have a parent class called Shape, which represents a generic shape and has a color attribute and a draw method.

The Circle class and Square class inherit from the Shape class. Each child class has its own unique attributes (radius for circles and side for squares) and methods (calculate_area).

We create instances of the child classes (circle and square) and demonstrate how they can access the methods from the parent class (draw), as well as their own unique methods (calculate_area).

The Circle class calculates the area of a circle based on its radius, while the Square class calculates the area of a square based on its side length.

This example showcases hierarchical inheritance, where the Shape class acts as the base for the Circle and Square classes, and each child class adds its own specific functionality.

# Multilevel Inheritance

Multilevel inheritance in Python refers to a scenario where a child class becomes the parent class for another child class. This creates a hierarchy of classes with each child class inheriting attributes and methods from its parent class and its grandparent class.

multilevel inheritance.JPG

```python
class Vehicle:
    def __init__(self, brand):
        self.brand = brand

    def drive(self):
        print(f"The {self.brand} vehicle is being driven.")


class Car(Vehicle):
    def __init__(self, brand, model):
        super().__init__(brand)
        self.model = model

    def honk(self):
        print(f"The {self.brand} {self.model} is honking.")


class ElectricCar(Car):
    def __init__(self, brand, model, battery_capacity):
        super().__init__(brand, model)
        self.battery_capacity = battery_capacity

    def charge(self):
        print(f"The {self.brand} {self.model} is charging. Battery capacity: {self.battery_capacity} kWh.")


# Create instances of the child classes
car = Car("Toyota", "Camry")
electric_car = ElectricCar("Tesla", "Model S", 85)

# Accessing methods from the parent classes
car.drive()  # Output: The Toyota vehicle is being driven.
electric_car.drive()  # Output: The Tesla vehicle is being driven.

# Accessing methods unique to each class
car.honk()  # Output: The Toyota Camry is honking.
electric_car.honk()  # Output: The Tesla Model S is honking.

# Accessing methods from the grandparent class
electric_car.charge()  # Output: The Tesla Model S is charging. Battery capacity: 85 kWh.
```

In this example, we have a parent class called Vehicle, which represents a generic vehicle and has a brand attribute and a drive method.

The Car class inherits from Vehicle and adds its own attributes (model) and methods (honk).

The ElectricCar class inherits from Car and further extends the functionality by adding its own attribute (battery_capacity) and method (charge).

We create instances of the child classes (car and electric_car) and demonstrate how they can access the methods from their respective parent classes (Vehicle and Car), as well as their own unique methods (honk for both and charge for ElectricCar).

This example illustrates multilevel inheritance, where the Car class becomes the parent for the ElectricCar class, creating a hierarchy of classes with each level adding its own attributes and methods.

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def eat(self):
        print(f"{self.name} is eating.")


class Mammal(Animal):
    def breathe(self):
        print(f"{self.name} is breathing.")


class Dog(Mammal):
    def bark(self):
        print(f"{self.name} is barking.")


# Create instances of the child classes
dog = Dog("Buddy")

# Accessing methods from the parent classes
dog.eat()  # Output: Buddy is eating.
dog.breathe()  # Output: Buddy is breathing.

# Accessing methods unique to the child class
dog.bark()  # Output: Buddy is barking.
```

In this example, we have a parent class called Animal, which represents a generic animal and has a name attribute and an eat method.

The Mammal class inherits from Animal and adds its own method (breathe).

The Dog class inherits from Mammal and further extends the functionality by adding its own method (bark).

We create an instance of the Dog class (dog) and demonstrate how it can access the methods from its parent classes (Animal and Mammal), as well as its own unique method (bark).

This example showcases multilevel inheritance, where the Mammal class becomes the parent for the Dog class, and the Animal class is the grandparent. Each level in the hierarchy adds its own attributes and methods, creating a chain of inheritance.

# Multiple Inheritance
Multiple inheritance in Python refers to a scenario where a child class can inherit from multiple parent classes. This allows the child class to inherit attributes and methods from all of its parent classes.

Multiple inheritance.JPG

```python
class Car:
    def __init__(self, brand):
        self.brand = brand

    def drive(self):
        print(f"The {self.brand} car is being driven.")


class ElectricVehicle:
    def __init__(self, battery_capacity):
        self.battery_capacity = battery_capacity

    def charge(self):
        print(f"The electric vehicle is charging. Battery capacity: {self.battery_capacity} kWh.")


class ElectricCar(Car, ElectricVehicle):
    def __init__(self, brand, battery_capacity):
        Car.__init__(self, brand)
        ElectricVehicle.__init__(self, battery_capacity)


# Create an instance of the child class
electric_car = ElectricCar("Tesla", 85)

# Accessing methods from the parent classes
electric_car.drive()  # Output: The Tesla car is being driven.
electric_car.charge()  # Output: The electric vehicle is charging. Battery capacity: 85 kWh.
```

In this example, we have two parent classes: Car and ElectricVehicle.

The Car class represents a generic car and has a brand attribute and a drive method.

The ElectricVehicle class represents a generic electric vehicle and has a battery_capacity attribute and a charge method.

The ElectricCar class inherits from both Car and ElectricVehicle using multiple inheritance. It can access attributes and methods from both parent classes.

We create an instance of the ElectricCar class (electric_car) and demonstrate how it can access the methods from both parent classes (Car and ElectricVehicle).

This example showcases multiple inheritance, where the ElectricCar class inherits from both the Car and ElectricVehicle classes, allowing it to combine their attributes and methods.

```python
class Person:
    def __init__(self, name):
        self.name = name

    def introduce(self):
        print(f"Hello, my name is {self.name}.")


class Employee:
    def __init__(self, emp_id):
        self.emp_id = emp_id

    def display_employee_info(self):
        print(f"Employee ID: {self.emp_id}")


class Manager(Person, Employee):
    def __init__(self, name, emp_id, department):
        Person.__init__(self, name)
        Employee.__init__(self, emp_id)
        self.department = department

    def manage_team(self):
        print(f"{self.name} is managing the {self.department} team.")


# Create an instance of the child class
manager = Manager("John", 1234, "Marketing")

# Accessing methods from the parent classes
manager.introduce()  # Output: Hello, my name is John.
manager.display_employee_info()  # Output: Employee ID: 1234.

# Accessing method unique to the child class
manager.manage_team()  # Output: John is managing the Marketing team.
```

In this example, we have two parent classes: Person and Employee.

The Person class represents a person and has a name attribute and an introduce method.

The Employee class represents an employee and has an emp_id attribute and a display_employee_info method.

The Manager class inherits from both Person and Employee using multiple inheritance. It can access attributes and methods from both parent classes.

We create an instance of the Manager class (manager) and demonstrate how it can access the methods from both parent classes (Person and Employee), as well as its own unique method (manage_team).

This example demonstrates multiple inheritance, where the Manager class inherits from both the Person and Employee classes, allowing it to combine their attributes and methods.

# Method Resolution Order (MRO) in Python

If two superclasses have the same method name and the derived class calls that method, Python uses the MRO to search for the right method to call. Python uses the C3 linearization algorithm to determine the MRO. The MRO ensures that classes are traversed in a consistent and predictable order, avoiding conflicts and maintaining the integrity of the inheritance hierarchy.

To view the MRO of a class, you can use the __mro__ attribute or the mro() method. Here's an example:

```python
class SuperClass1:
    def info(self):
        print("Super Class 1 method called")

class SuperClass2:
    def info(self):
        print("Super Class 2 method called")

class Derived(SuperClass1, SuperClass2):
    pass

d1 = Derived()
d1.info()

# Output: "Super Class 1 method called"
```

Here, SuperClass1 and SuperClass2 both of these classes define a method info().

So when info() is called using the d1 object of the Derived class, Python uses the MRO to determine which method to call.

In this case, the MRO specifies that methods should be inherited from the leftmost superclass first, so info() of SuperClass1 is called rather than that of SuperClass2.

```python
class A:
    def some_method(self):
        pass

class B(A):
    def some_method(self):
        pass

class C(A):
    def some_method(self):
        pass

class D(B, C):
    def some_method(self):
        pass

print(D.__mro__)
# Output: (<class '__main__.D'>, <class '__main__.B'>, <class '__main__.C'>, <class '__main__.A'>, <class 'object'>)

print(D.mro())
# Output: [<class '__main__.D'>, <class '__main__.B'>, <class '__main__.C'>, <class '__main__.A'>, <class 'object'>]
```

In this example, we have four classes: A, B, C, and D. The class D inherits from both B and C, which in turn inherit from A.

By printing the __mro__ attribute or using the mro() method on the D class, we can see the MRO. In this case, the MRO of class D is D, B, C, A, and object. This means that when resolving a method or attribute in D, Python will first look in D, then B, then C, then A, and finally in the built-in object class.

Understanding the MRO is important in cases of multiple inheritance, as it determines the order in which methods and attributes are inherited and resolved in the class hierarchy.

```python
class A:
    def some_method(self):
        print("Method from class A")

class B(A):
    def some_method(self):
        print("Method from class B")
        super().some_method()

class C(A):
    def some_method(self):
        print("Method from class C")
        super().some_method()

class D(B, C):
    def some_method(self):
        print("Method from class D")
        super().some_method()

# Create an instance of the child class
d = D()

# Call the method
d.some_method()
```

The method resolution follows the MRO, starting from the class D, then moving to B, then C, and finally to A. The super().some_method() calls the method of the immediate parent class in the MRO, resulting in the printed statements above.

This example demonstrates the MRO and how it determines the order of method resolution in multiple inheritance.

# Method Overriding in Python Inheritance

what if the same method is present in both the superclass and subclass?

In this case, the method in the subclass overrides the method in the superclass. This concept is known as method overriding in Python.

```python
class Animal:

    # attributes and method of the parent class
    name = ""

    def eat(self):
        print("I can eat")

# inherit from Animal
class Dog(Animal):

    # override eat() method
    def eat(self):
        print("I like to eat bones")

# create an object of the subclass
labrador = Dog()

# call the eat() method on the labrador object
labrador.eat()
```

In the above example, the same method eat() is present in both the Dog class and the Animal class.

Now, when we call the eat() method using the object of the Dog subclass, the method of the Dog class is called.

This is because the eat() method of the Dog subclass overrides the same method of the Animal superclass.

# The super() Method in Python Inheritance

Previously we saw that the same method in the subclass overrides the method in the superclass.

However, if we need to access the superclass method from the subclass, we use the super() method. For example,

```python
class Animal:

    name = ""

    def eat(self):
        print("I can eat")

# inherit from Animal
class Dog(Animal):

    # override eat() method
    def eat(self):

        # call the eat() method of the superclass using super()
        super().eat()

        print("I like to eat bones")

# create an object of the subclass
labrador = Dog()

labrador.eat()
```

In the above example, the eat() method of the Dog subclass overrides the same method of the Animal superclass.

Inside the Dog class, we have used
# call method of superclass
super().eat()
to call the eat() method of the Animal superclass from the Dog subclass.

So, when we call the eat() method using the labrador object.
# call the eat() method
labrador.eat()
Both the overridden and the superclass version of the eat() method is executed.

```python
class Vehicle:
    def start(self):
        print("Vehicle started.")


class Car(Vehicle):
    def start(self):
        super().start()  # Call the start() method of the parent class
        print("Car started.")


# Create an instance of the child class
car = Car()

# Call the overridden method
car.start()
```

In this example, we have a parent class called Vehicle with a start() method that prints "Vehicle started."

The Car class inherits from Vehicle and overrides the start() method. Inside the overridden method, we use super().start() to invoke the start() method of the parent class before extending it.

When we create an instance of the Car class and call the start() method, it first invokes the start() method of the parent class using super().start(), which prints "Vehicle started." Then, it continues to execute the code in the overridden method and prints "Car started."

By using the super() function, you can leverage the functionality of the parent class while extending or modifying it in the child class. It allows you to create more flexible and maintainable code in an inheritance hierarchy
